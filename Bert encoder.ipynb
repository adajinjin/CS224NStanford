```python
from bert_serving.client import BertClient
import numpy as np
# from termcolor import colored
```

```python
bc = BertClient(ip='10.42.18.124') #for remote server
```

```python
bc.encode(['First do it', 'then do it right', 'then do it better'])
```

```python
bc.encode(['First do it', 'then do it right', 'then do it better']).shape
```

```python
text1 = 'I love cooking.'
text2 = 'cooking is my favorite activity.'
text3 = 'Cooking seems pretty interesting to me.'
text4 = 'I think this method is the most commonly used for customer satisfaction survey'
```

```python
textencode = bc.encode([text1, text2, text3, text4])
questions = [text1, text2, text3, text4]
```

```python
def find_similarity(query_question, topk=1):
    query = bc.encode([query_question])
    score = np.sum(query * textencode, axis=1) / np.linalg.norm(query, axis=1)
    topk_idx = np.argsort(score)[::-1][:topk]
    for idx in topk_idx:
        print('> %s\t%s' % ('%.1f' % score[idx], questions[idx]))
```

```python
# query_question = 'Playing football is my favorite activity.'
# query_question = 'I enjoy cooking very much.'
query_question = 'For customer satisfaction, I would want to use this survey template'
```

```python
find_similarity(query_question)
```
